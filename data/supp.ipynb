```python
from random import randint 
import pandas as pd 
import random
import numpy as np
from faker import Faker
 
fake1 = Faker('Polish')
fake = Faker()

 
def input_data(x):

    # pandas dataframe
    data = pd.DataFrame()
    for i in range(0, x):
        data.loc[i,'supplier_id']= str(i)
        data.loc[i,'company']= fake.company()
        data.loc[i,'name']= fake.name()
        data.loc[i,'address']= str(fake.address().replace("\n", " "))
        data.loc[i,'nip']= str(random.randrange(1111111111, 9999999999, 10))
        data.loc[i,'phone_number']= str(fake1.phone_number())
        data.loc[i,'email']= str(fake.email())
    data['email']=data['email'].str.slice(0,50)
    data['address']=data['address'].str.slice(0,50)
    return data
   
   
supp=input_data(30)
print(supp)
supp.to_csv("C:\\Users\\s1z2y\\Desktop\\pythonlogin\\data\\suppliers.csv",index=False)
```
